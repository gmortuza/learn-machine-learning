# Overfitting
Overfitting refers to a model that models the training data too well.
Overfitting happens when a model **learns the detail and noise** in the training data to the extent that it negatively impacts the performance of the model on new data. This means that the noise or random fluctuations in the training data is picked up and learned as concepts by the model. The problem is that these concepts do not apply to new data and negatively impact the models ability to generalize. More likely to happen in **nonparametric and nonlinear models**. For example, decision trees are a nonparametric machine learning algorithm that is very flexible and is subject to overfitting training data. 

We can solve overfitting using:
- Regularization
- Cross Validation
- Drop out
- Early Stoping (Downside orthogonalization)
